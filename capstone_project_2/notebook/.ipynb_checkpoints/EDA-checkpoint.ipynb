In the next step we will visulize and test the association between various categorical features and our taret feature TenYearCHD. We will visulaize the data as bar graph and test the association using Chi-square test of indepenece.

```python
#function to prepare crosstab of selected features and perform chisqure test
def chi2(var1, var2, dataframe):
    """
    Prepare crosstab of selected columns and perform chisquare test
    """
    ctab = pd.crosstab(dataframe[var1], dataframe[var2])
    chi_val, p_val, dof, exp = chi2_contingency(ctab)
    percent_ctab = 100*ctab/ctab.sum()
    return percent_ctab, chi_val, p_val
```

#####  Association between gender and TenYearCHD

```python
#Association between gender and CHD
df1, chisq, pval = chi2('TenYearCHD', 'gender', df)

gender = df1.rename(columns={0: 'Female', 1: 'Male'}, index={0: 'No CHD', 1: 'CHD'})
print(gender)
gender.T.CHD.plot.bar(color = 'teal')
plt.ylabel('Percentaage')
plt.title('CHD by gender')
plt.show()
```

```python
#Association between gender and CHD
df1, chisq, pval = chi2('TenYearCHD', 'gender', df)

gender = df1.rename(columns={0: 'Female', 1: 'Male'}, index={0: 'No CHD', 1: 'CHD'})
print(gender)
gender.T.CHD.plot.bar(color = 'teal')
plt.ylabel('Percentaage')
plt.title('CHD by gender')
plt.show()
```

```python
#Chi-square test of independence
print("Chisquare value: ", chisq, "P-value: ", pval)
```

Analysis showed that 12.28% of the sampled female had CHD while this number increased to 18.91% in male. Chi-square test also showed that there is relationship between gender and CHD.

# EDA of numeric data

```python
#list of numberical data
num = ['cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']
```

```python
#histogram demonstrating distribution of numerical features 
x = 2  # subplot row number
y = 4  # subplot column number
z = 1  # counter

fig = plt.figure(figsize=(16,8))
fig.subplots_adjust(wspace=0.5, hspace=0.3)

for i in num:
    plt.subplot(x, y, z)
    plt.title(i)
    plt.xlabel(i)
    sns.histplot(df[i][df['TenYearCHD']==0], color = 'blue', alpha = 0.5)
    sns.histplot(df[i][df['TenYearCHD']==1], color = 'red', alpha = 0.5)
    plt.axvline(np.mean(df[i][df['TenYearCHD']==0]), color = 'blue', linestyle ='dashed', linewidth =1)
    plt.axvline(np.mean(df[i][df['TenYearCHD']==1]), color = 'red', linestyle ='dashed', linewidth =1)
    fig.legend(labels=['No CHD','CHD'])
    z += 1
    
plt.show()
```

```python
# Box plot showing the comparision of the distribution of various numerical data across samples with or without CHD.
x = 2  # subplot row number
y = 4  # subplot column number
z = 1  # counter

fig = plt.figure(figsize=(16,8))
fig.subplots_adjust(wspace=0.5, hspace=0.3)

for i in num:
    plt.subplot(x, y, z)
    plt.title(i)
    plt.xlabel(i)
    sns.boxplot(y=i, x='TenYearCHD', data=df)
    z += 1

plt.show()
```

```python
#Function to perform t-test
def t_test(D1, D2, alpha):
    """
    function to perform t-test
    """
    stat, p = ttest_ind(D1, D2)
    print('Statistics=%.3f, p=%.3f' % (stat,p))
    if p> alpha:
        print('fail to reject H0')
    else:
        print('reject H0')
```

```python
t_test(df['cigsPerDay'][df['TenYearCHD']==0], df['cigsPerDay'][df['TenYearCHD']==1] ,0.05)
```

```python
t_test(df['totChol'][df['TenYearCHD']==0], df['totChol'][df['TenYearCHD']==1] ,0.05)
```

```python
t_test(df['sysBP'][df['TenYearCHD']==0], df['sysBP'][df['TenYearCHD']==1] ,0.05)
```

```python
t_test(df['diaBP'][df['TenYearCHD']==0], df['diaBP'][df['TenYearCHD']==1] ,0.05)
```

```python
t_test(df['BMI'][df['TenYearCHD']==0], df['BMI'][df['TenYearCHD']==1] ,0.05)
```

```python
t_test(df['heartRate'][df['TenYearCHD']==0], df['heartRate'][df['TenYearCHD']==1] ,0.05)
```

```python
t_test(df['glucose'][df['TenYearCHD']==0], df['glucose'][df['TenYearCHD']==1] ,0.05)
```

t-test showed that except heart rate, sampled population with CHD had sginificantly higher mean for 'cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', and 'glucose' than the sampled population without CHD.
