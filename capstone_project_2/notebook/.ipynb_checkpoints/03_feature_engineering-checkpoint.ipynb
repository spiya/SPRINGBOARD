# Feature engineering

```python
#import necessary library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import RobustScaler
```

```python
# Read the datafile
df = pd.read_csv('EDA_cleaned_CHD.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe().T
```

```python
#convert numerical data type 'age' into catergorical data as age group
df['age_group'] = pd.cut(df['age'], bins=[1, 30, 40, 50, 60, np.inf], 
                          labels=[1,2,3,4,5])
df.drop(['age'], axis=1, inplace=True)
df.head()
```

# Check multicollinearity

```python
#import the package
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# defining the independent variables
X = df.drop('TenYearCHD', axis=1)

# add constant to the independent variables
X = sm.add_constant(X)

# defining the dependent variables
y = df[['TenYearCHD']]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
  
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
  
print(vif_data)
```

```python
# building the model and fitting the data
log_reg = sm.Logit(y, X).fit()
```

```python
# printing the summary table
print(log_reg.summary())
```

```python
# feature scaling
R_scaler = RobustScaler()
X_scaler = R_scaler.fit_transform(df.drop('TenYearCHD', axis=1))
X=pd.DataFrame(X_scaler, columns=df.drop('TenYearCHD', axis=1).columns)
X = sm.add_constant(X)
X.head()
```

```python
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
  
print(vif_data)
```

```python
# building the model and fitting the data
log_reg = sm.Logit(y, X).fit()
```

```python
# printing the summary table
print(log_reg.summary())
```

```python
logreg_coef = pd.concat([pd.DataFrame(log_reg.params.index),pd.DataFrame(log_reg.params.values)],axis=1)
logreg_coef.columns = ['Features','coef']
logreg_coef.sort_values("coef", ascending=False, inplace=True)
logreg_coef
```

```python
#Bar plot of feature importance
sns.barplot(x='coef', y='Features', data = logreg_coef)
plt.xlabel('coefficient')
plt.title('Feature importance')
plt.show()
```

## Identify the features that contribute the most to coronary heart disease

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=['TenYearCHD'])
y = df.TenYearCHD
#scale the features
R_scaler = RobustScaler()
X_scaler = R_scaler.fit_transform(X)
X=pd.DataFrame(X_scaler, columns=X.columns)
```

```python
## Import the random forest model.
from sklearn.ensemble import RandomForestClassifier
## This line instantiates the model.
rf = RandomForestClassifier()
## Fit the model on your training data.
rf.fit(X_scaler, y)
```

```python
feature_importances = pd.DataFrame(rf.feature_importances_,
                        index = X.columns,
                        columns=['importance']).sort_values('importance', ascending=False)
feature_importances
```

```python
#Bar plot of feature importance
sns.barplot(x=feature_importances.importance, y=feature_importances.index)
plt.xlabel('Scores of features')
plt.title('Feature importance')
plt.show()
```

## Select the top 10 most important features to predict CHD

```python
# select the top 10 most important features
top_10_features = feature_importances.index.tolist()[:10]
top_10_features
```

```python
new_df = pd.concat([X[top_10_features],pd.DataFrame(y)], axis=1)
new_df.head()
```

```python
#save the new dataframe
new_df.to_csv(r"03_FE_CHD_data.csv",  index = False, header=True)
```
